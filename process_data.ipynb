```python
messages = df.message
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem.wordnet import WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer
```

```python
stop_words = stopwords.words('english')
lemmatizer = WordNetLemmatizer()
```

```python
#tokens = [word_tokenize(sent) for sent in messages]
```

```python
def tokenize(text):
    text = re.sub(r'[^a-zA-Z0-9]', " ", text.lower())
    
    tokens = word_tokenize(text)
    
    tokens_nostop_words = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]
    
    return tokens_nostop_words
```

```python
vect = CountVectorizer(tokenizer=tokenize)
```

```python
X = vect.fit_transform(messages)
```

```python
list((vect.vocabulary_).keys()).__len__()
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
transformer = TfidfTransformer(smooth_idf=False)
```

```python
tfidf = transformer.fit_transform(X)
```
